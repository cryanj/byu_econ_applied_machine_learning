```python
sns.lmplot(x="pepsi", y="pr_pepsi", data=soda_data)
rx = plt.gca()
rx.set_title("Demand Curve of Pepsi")
```

```python
sns.lmplot(x="sevenup", y="pr_7up", data=soda_data)
rx = plt.gca()
rx.set_title("Demand Curve of 7Up")
```

```python
sns.lmplot(x="coke", y="pr_coke", data=soda_data)
rx = plt.gca()
rx.set_title("Demand Curve of Coke")
```

This is actually super cool! Here are the demand curves for each soda relative to the other sodas.  They are all highly elastic; a small change inpric leads to a big change in quantity bought.  Who would have ever thought that pepsi although still elastic would be less elastic than the other sodas!?  Really Pepsi?!
